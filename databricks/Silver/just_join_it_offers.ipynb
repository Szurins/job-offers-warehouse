```python
from pyspark.sql import functions as fs
```

# Loading Just Join It Offers

```python
df = spark.table("`job-offers-warehouse`.bronze.justjoinit_job_offers")
```

```python
# Column "category" does not need any cleaning
df.groupBy(fs.col("category")).count().orderBy(fs.col("count")).display()
```

```python
# Column "city" needed data standardization (Krakow and Kraków are now the same thing)

city_map = {
    # International Translations
    "Warszawa": "Warsaw", "WARSAW": "Warsaw", "warsaw": "Warsaw", 
    "Warszawa (Mazowieckie)": "Warsaw", "Warsaw Hybrid": "Warsaw", 
    "Warsaw Onsite": "Warsaw", "Warsaw, Poland": "Warsaw",
    "Krakow": "Kraków", "Cracow": "Kraków",
    "Londyn": "London",
    "Bruksela": "Brussels",
    "Budapeszt": "Budapest",
    "Bukareszt": "Bucharest", "Bucharest": "Bucharest",
    "Cologne": "Köln",
    "Copenhagen": "København", "København V": "København",
    "Lisboa": "Lisbon", "Lisbona": "Lisbon",
    "Praha": "Prague",
    "Rīga": "Riga",
    "Nikosia": "Nicosia",
    "Lublana": "Ljubljana",
    "Nowy Jork": "New York",
    "Abu Zabi": "Abu Dhabi",
    "Luksemburg": "Luxembourg", "Luxemburg": "Luxembourg",
    
    # Misspellings & Variants
    "Gydnia": "Gdynia",
    "Vellentuna": "Vallentuna",
    "Hanover": "Hannover",
    "Bad Homburg Vor Der Höhe": "Bad Homburg vor der Höhe",
    "Stockholm City": "Stockholm", "Stockholm Metropolitan Area": "Stockholm",
    "Madrid hybrid": "Madrid",
    
    # Country/Broad Logic
    "Arabia Saudyjska": "Saudi Arabia",
    "polska": "Poland", "Polska": "Poland", 
    "Poland (Remote)": "Poland", "Poland remote": "Poland"
}
df = df.replace(to_replace=city_map, subset=['city'])
df.groupBy(fs.col("city")).count().orderBy(fs.col("city")).display()
```

```python
# Column "workplaceType" does not need any cleaning
df.groupBy(fs.col("workplaceType")).count().display()
```

```python
# Column "niceToHaveSkills" does not need any cleaning
df.groupBy(fs.col("niceToHaveSkills")).count().display()
```
